```python
class Pet:
    
    #initialize the instance variables
    def __init__(self,n,a,w,pr):
        self.__petname = n
        self.__age = a
        self.__weight = w
        self.__person = pr

    #initialize the acessor methods
    def get_petname(self):
        return self.__petname
    
    def get_age(self):
        return self.__age
    
    def get_weight(self):
        return self.__weight
    
    #initialize the mutator methods
    def set_petname(self):
        self.__petname = n
        
    def set_age(self):
        self.__age = a
        
    def set_weight(self):
        self.__weight = w

        
    def __str__(self):
        st = "Pet Name: {}; Pet Age: {}; Pet Weight: {};".format(self.__petname,self.__age,self.__weight) 
        st += "\n" + self.__person.__str__()
        return st
        

        
```

```python
class Mammals(Pet):
    
    #initialize instance variables
    def __init__(self,n,a,w,pr,l,c):
        super().__init__(n,a,w,pr)
        self.__name = n
        self.__age = a
        self.__weight = w
        self.__littersize = l
        self.__hasclaws = c
        
    
    #initialize the acessor methods
    def get_littersize(self):
        return self.__littersize
    
    def get_hasclaws(self):
        return self.__hasclaws    
    
    #initialize the mutator methods
    def set_littersize(self):
        self.__littersize = l
        
    def set_hasclaws(self):
        self.__hasclaws = c        
        
    def __str__(self):
        stm = super().__str__()
        stm += "\n Litter Size: {}; Has Claws: {}".format(self.__littersize,self.__hasclaws)
        return stm
                
    def weight_gain(self):
        if self.__age < 60:
            return "{} is too young to have added weight".format(self.__name)
        else:
            if self.__age < 300:
                wg = ((self.__age // 60) * (0.1 * self.__weight))+ self.__weight
            else:
                wg = (5 * (0.1 * self.__weight))+ self.__weight
            return "{}'s weight gain is {}".format(self.__name,wg)
        
```

```python
class Fish(Pet):
    
    #initialize instance variables
    def __init__(self,n,a,w,pr,sc,le):
        super().__init__(n,a,w,pr)
        self.__name = n
        self.__age = a
        self.__weight = w
        self.__scalecondition = sc
        self.__length = le
        
    
    #initialize the acessor methods
    def get_scalecondition(self):
        return self.__scalecondition
    
    def get_length(self):
        return self.__length   
    
    #initialize the mutator methods
    def set_scalecondition(self):
        self.__scalecondition = sc
        
    def set_length(self):
        self.__length = le        
        
    def __str__(self):
        stf = super().__str__()
        stf += " \n Scale Condition: {}; Length: {}".format(self.__scalecondition,self.__length)
        return stf
    
    def weight_gain(self):
        if  self.__age < 0:
            return "{} is too young to have gained weight".format(self.__name)
        else:
            if self.__age < 240:
                wg = ((self.__age // 80) * (0.05 * self.__weight))+ self.__weight
            else:
                wg = (3 * (0.05 * self.__weight))+ self.__weight
            return "{}'s weight gain is {}".format(self.__name,wg)
        
        
```

```python
class Amphibian(Pet):
    
    #initialize instance variables
    def __init__(self,n,a,w,pr,v):
        super().__init__(n,a,w,pr)
        self.__name = n
        self.__age = a
        self.__weight = w
        self.__isvenomous = v
     
    #initialize the acessor methods
    def get_isvenomous(self):
        return self.__isvenomous
        
    #initialize the mutator methods
    def set_isvenomous(self):
        self.__isvenomous = v
            
    def __str__(self):
        stv = super().__str__()
        stv += " \n is Venomous: {} ".format(self.__isvenomous)
        return stv
    

    def weight_gain(self):
        if self.__age < 120:
            return "{} is to young too have gain weight".format(self.__name)
        else:
            if self.__age < 360:
                wg = ((self.__age // 120) * (0.03 * self.__weight))+ self.__weight
            else:
                wg = (3 * (0.03 * self.__weight))+ self.__weight
            return "{}'s weight gain is {}".format(self.__name,wg)
        
```

```python
def petoutput(pet, n):
    if n == 1:
        with open('petdata.dat', 'ab') as f:
             pickle.dump(pet,f)
    else:
        with open('petweight.dat', 'ab') as f:
             pickle.dump(pet,f)
    
```
